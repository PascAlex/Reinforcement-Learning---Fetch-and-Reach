
Import all libraries

```python
import gym
import numpy as np
from stable_baselines3.ddpg.policies import MultiInputPolicy
from stable_baselines3.common.noise import OrnsteinUhlenbeckActionNoise, NormalActionNoise
from stable_baselines3 import DDPG, TD3, SAC, PPO
```

Create the environmnet

```python
env = gym.make('FetchReach-v1')
```

Create the model for all the algorithms

```python
#number of timesteps
n_timesteps = 300000
#log interval
log_interval = 4
#tensorboar log
log="./fetchreachGraphs/"

# the noise objects for DDPG and TD3
n_actions = env.action_space.shape[-1]

#action_noise = OrnsteinUhlenbeckActionNoise(mean=np.zeros(n_actions), sigma=float(0.5) * np.ones(n_actions))
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))

modelDDPG = DDPG(MultiInputPolicy,
             env, verbose=1,
             action_noise=action_noise,
             tensorboard_log=log)
modelDDPG.learn(total_timesteps=n_timesteps,log_interval=log_interval)
modelDDPG.save("DDPG_fetch_reach_with_graphics")

modelTD3 = TD3("MultiInputPolicy",
            env, verbose=1,
            action_noise=action_noise,
            tensorboard_log=log)
modelTD3.learn(total_timesteps=n_timesteps, log_interval=log_interval)
modelTD3.save("TD3_fetch_reach_with_graphics")

modelPPO = PPO("MultiInputPolicy",env,
               verbose=1,n_steps=200,
               tensorboard_log=log)
modelPPO.learn(total_timesteps=n_timesteps)
modelPPO.save("PPO_fetch_reach_with_graphics")

modelSAC = SAC("MultiInputPolicy", env,
               verbose=1,learning_rate=0.001,
               learning_starts = 1000,
               gamma = 0.95,
               tensorboard_log=log)
modelSAC.learn(total_timesteps=n_timesteps, log_interval=log_interval)
modelSAC.save("SAC_fetch_reach_with_graphics")
```

In order to see how the results are performing on the environment run the
following line of code

```python
model = DDPG.load("DDPG_fetch_reach_with_graphics.zip")
obs =env.reset()


for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        action, _states = model.predict(obs)
        obs, rewards, dones, info = env.step(action)
        if dones:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
